### Convert fare in indian rupees and create new column

```python
df.drop(['new_column', 'id+Pclass'], axis=1, inplace=True)
```

#### apply function
applu function similary work like a map function

```python
df["Fare in rupees"] = df["Fare"].apply(lambda x : x*87.96)
```

```python
df
```

#### make a fare category

```python
def fare_caterogry(x):
    category = ""
    if x < 20:
        category = "Cheaper"
    elif x >= 20 and x < 40:
        category = "Medium"
    elif x >=40:
        category = "Higher"

    return category
```

```python
df["Fare category"] = df["Fare"].apply(fare_caterogry)
```
